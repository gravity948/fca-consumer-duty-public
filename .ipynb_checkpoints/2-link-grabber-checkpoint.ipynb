```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from usp.tree import sitemap_tree_for_homepage
from urllib.parse import urljoin
import urllib.request
import ast
headers = {"User-Agent": "Mozilla/5.0"}
cookies = {"CONSENT": "YES+cb.20210720-07-p0.en+FX+410"}
```

```python
def readin_urls():
    firm_df = pd.read_excel('Company Names.xlsx', sheet_name='Company Names')
    firm_urls_dic = {}
    try:
        firm_urls_dic = dict(pd.read_excel("urls.xlsx").values)
    except:
        print("error reading in list of urls, either list doesnt exist or bad read")
```

```python
for firm in firm_df["Name"]:
    if firm in firm_urls_dic:
        continue
    print(f"{len(firm_urls_dic)+1:<5} {firm} ", end=' ')
    results = 100 # valid options 10, 20, 30, 40, 50, and 100
    page = requests.get(f"https://www.google.com/search?q={firm}",cookies=cookies)
    soup = BeautifulSoup(page.content, 'html.parser')
    links = soup.findAll("a")
    firm_urls = {}
    for link in links:
        link_href = link.get('href')
        if "/url?q=" in link_href:
            #this is kinda dodgy, but url will have 3 / before the www so element 3 of the array will solely be the domain
            firm_url = (link_href.split("/"))[3]
            print(firm_url)
            if firm_url is not None and not(('gov.uk' or 'linkedin' or 'bloomberg')  in firm_url) :
                    firm_urls_dic[firm]=firm_url
                    break
```

```python
def dic_to_excel(dic, name):
    output_df = pd.DataFrame(dic.items(), columns=["Firm","URL"])
    output_df.to_excel(name,index=False)
```

```python
def url_checker(url):
    try:
        get = requests.get(url)
        # if the request succeeds 
        if get.status_code == 200:
            return True
        else:
            return False
    #Exception
    except requests.exceptions.RequestException as e:
        return False
```

```python
firm_df_clean = pd.read_excel('Clean_Firm_Urls.xlsx')
temp_df1 = pd.read_excel('TandC.xlsx')
temp_df2 = pd.read_excel("TandC_latest.xlsx")
for index, row in temp_df2.iterrows():
    firms_tandc[row["Firm"]] = row["URL"]
for index, row in firm_df_clean.iterrows():
    if index % 10 ==0:
        print("-------------------------saved")
        dic_to_excel(firms_tandc, "TandC.xlsx")
        
    firm = row["Firm"]
    print(firm)
    if(firm in firms_tandc):
        print('in dic already')
        continue
    

    url = "https://"+row["URL"]
    print(url)
    if(not url_checker(url)):
        continue
    try:
        reqs = requests.get(url)
    except:
        continue
    soup = BeautifulSoup(reqs.text, 'html.parser')

    urls = []

    for link in soup.find_all('a'):
        if link.get("href"):
            # if the link tag has the 'href' attribute
            css_url = urljoin(url, link.get("href"))
            urls.append(css_url)

    substr = ['terms-and-conditions', 'privacy-and-cookies', 't&c', 'privacy', 'cookies', 'terms', 'cookie']
    x=[]
    for i in substr:
        for url in urls:
            if(url.find(i)!=-1 and url not in x):
                x.append(url)
    print(x)
    firms_tandc[firm] = x
```
